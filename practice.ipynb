## üß™ Python Demo Script: `noreturn_vs_none_demo.py`

```python
from typing import NoReturn
import sys

# ‚úÖ Proper NoReturn: program yahin khatam
def shutdown() -> NoReturn:
    print("‚ö†Ô∏è Shutting down system...")
    sys.exit()

# ‚úÖ Proper NoReturn: infinite loop, kabhi return nahi karega
def loop_forever() -> NoReturn:
    while True:
        print("üîÅ Still running...")
        # Note: isme break nahi hona chahiye

# ‚úÖ This returns None (default behavior)
def greet():
    print("üëã Hello! I'm a normal function.")

# üëá Try calling one by one

# greet()         # Output: üëã Hello!
# shutdown()      # Output: ‚ö†Ô∏è Shutting down system... and exit
# loop_forever()  # Output: üîÅ Still running... (infinite)
```

---

## üß† Important Tips:

| Function         | Return Type | Description                                |
|------------------|-------------|--------------------------------------------|
| `greet()`        | `None`      | Normal function, koi return nahi diya      |
| `shutdown()`     | `NoReturn`  | Program ko turant band karta hai           |
| `loop_forever()` | `NoReturn`  | Kabhi rukta hi nahi (infinite loop)        |

## üß† **Use Case: Emergency Monitoring App**

### ü©∫ App ka kaam:
- Background me continuously sensors check karta hai (heart rate, temperature etc.)
- Agar kuch galat ho to program **immediate shutdown** karta hai (`sys.exit()`)
- Normal checks ke liye `None` return hota hai
- Kabhi kabhi program ko **kabhi band nahi hona hota** (jaise background monitoring) ‚Üí `NoReturn` + `while True`

---

### ‚úÖ Architecture Overview

| Function Name     | Kaam kya karta hai                        | Use Kya Hua?         |
|-------------------|-------------------------------------------|-----------------------|
| `monitor_loop()`  | Hamesha sensor monitor karta rahta hai    | `NoReturn` + `while True` |
| `shutdown_system()` | Agar critical alert aaye to system band  | `sys.exit()` + `NoReturn` |
| `send_report()`   | Report bhejta hai aur `None` return karta | Returns `None`         |

---

## ‚úÖ Sample Code:

```python
from typing import NoReturn
import sys
import time

# üì§ Ye sirf ek kaam karke khatam ho jata hai
def send_report():
    print("üì§ Report sent to the server.")
    # No return likha nahi, lekin default is None
    return

# üõë Agar kuch serious alert aaye to system band karo
def shutdown_system() -> NoReturn:
    print("üö® Critical Alert! Shutting down system...")
    sys.exit()  # Poora program yahin ruk jaata hai

# üîÅ Ye background monitor karta rahega kabhi rukta nahi
def monitor_loop() -> NoReturn:
    while True:
        print("ü©∫ Monitoring sensors...")
        # Dummy condition
        heart_rate = 120
        if heart_rate > 100:
            shutdown_system()  # Poora system band
        time.sleep(5)

# ‚úÖ Start the app
# send_report()        # Normal call
monitor_loop()         # Ye kabhi khatam nahi hoga
```

---

## ‚úÖ Real-life Mapping:

| Concept | Real Use |
|--------|-----------|
| `None` | Jab function sirf kaam karke wapis aa jaye (like report send, logging) |
| `sys.exit()` | Jab kuch serious ho jaaye (security breach, system failure) |
| `NoReturn` | Jab function ya to `sys.exit()` kare ya loop se kabhi return na kare |
| `while True` | Jab koi **background process ya infinite task** ho |
