## 🔧 1. **CPU (Central Processing Unit)**

> Ye tumhare computer ka **main brain** hota hai.

### ⚙️ Use:
- Normal daily tasks ke liye: browser chalana, Word file open karna, coding, etc.
- Ek waqt me **thode se tasks** kar sakta hai (usually 4 ya 8 cores).

### ✅ Achha:
- Har tarah ka kaam handle kar sakta hai.
- Flexible hota hai.

### ❌ Limit:
- **Slow** hota hai jab bohot zyada data ya parallel task ho, jaise AI training.

---

## 🎮 2. **GPU (Graphics Processing Unit)**

> Ye originally **games aur graphics** ke liye bana tha, lekin ab **AI aur machine learning** me bhi use hota hai.

### ⚙️ Use:
- **Deep learning, neural networks, image processing**
- Ek waqt me **hazaron chhoti calculations** kar sakta hai (parallel processing).

### ✅ Achha:
- **Fast** hota hai ML models train karne ke liye.
- Multiple cores hotay hain (1000+).

### ❌ Limit:
- General tasks ke liye utna efficient nahi jitna CPU.

---

## 🤖 3. **TPU (Tensor Processing Unit)**

> Ye Google ne specially **TensorFlow aur AI training** ke liye banaya hai.

TensorFlow ek open-source software library hai jo Google ne banai hai,
taake hum machine learning (ML) aur deep learning (AI) ke models asaani se bana sakein aur train kar sakein.

### ⚙️ Use:
- Sirf **AI aur machine learning tasks** ke liye.
- TensorFlow models ko **super speed** se train karta hai.

### ✅ Achha:
- GPU se bhi **fast** hota hai specific AI tasks ke liye.
- Google Cloud pe free ya paid milta hai (Colab me bhi limited access hota hai).

### ❌ Limit:
- Sirf AI-related kaam ke liye bana hai.
- General purpose computing nahi kar sakta.

---

## 🔍 Ek Table me Summary:

| Feature        | CPU              | GPU                  | TPU                         |
|----------------|------------------|-----------------------|------------------------------|
| Use            | General purpose  | Graphics + AI         | Sirf AI/ML (TensorFlow)     |
| Speed          | Slowest          | Fast                  | Fastest (for TensorFlow)    |
| Cores          | Few (4-16)       | Many (1000+)          | Specialized matrix cores     |
| Flexibility    | Har kaam ke liye | Mostly parallel tasks | Sirf TensorFlow tasks        |
| Maker          | Intel, AMD       | NVIDIA, AMD           | Google                       |
